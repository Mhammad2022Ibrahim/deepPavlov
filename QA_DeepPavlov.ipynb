```python
from deeppavlov.core.commands.utils import parse_config

model_config = parse_config('qa_squad2_bert')
```

```python
from deeppavlov import build_model, configs
model = build_model(configs.squad.qa_squad2_bert, download=True)
```

```python
import pandas as pd
data = pd.read_csv("jokes.csv")
data = data.rename(columns={"Question": "questions", "Answer": "answers"})
data.head()
```

```python
questions = data.questions.to_list()
questions
```

```python
import random

def answer_question(question, data):
    # Check if the question is in the data DataFrame
    if question in data['questions'].values:
        relevant_question = question
    else:
        # If the question is not in the DataFrame, select a random question
        relevant_question = data['questions'].sample(n=1).values[0]
        
    # Find the most relevant answer from the jokes dataset
    relevant_answer = data.loc[data['questions'] == relevant_question, 'answers'].iloc[0]
    
    # Use the deeppavlov model to extract the answer from the relevant text
    extracted_answer = model([relevant_answer], [relevant_question])[0][0]
    
    return extracted_answer, relevant_question
```

```python
while True:
    # Ask the user to input a question
    user_question = input("Ask a question: ")
    if user_question.lower() in ['bye', 'goodbye']:
      print("Chatbot: Goodbye!")
      break
    # Answer the user question
    answer, relevant_question = answer_question(user_question, data)

    # Print the answer and relevant question
    print(f"Q: {relevant_question}")
    print(f"A: {answer}")
    print()
```
